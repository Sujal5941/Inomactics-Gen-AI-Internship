### 1) mobile recharge validation system.

```python
#### This program validaets recharge amount and plan. Aloows retry if invalid.

# Function to validate recharge details
def validate_recharge(amount,plan):
  valid_plans = [199,299,399,599]

  # Check minimum recharge amount
  if amount < 50:
    print("Recharge amount must be at least ₹50.")
    return False
  # Check if selected plan is valid
  if plan not in valid_plans:
    print("Invlaid plan selected.")
    return False

  return True

# While loop to allow retry
while True:
  amount = int(input("Enter recharge amount: "))
  plan = int(input("Enter plan (199/299/399/599): "))

  if validate_recharge(amount, plan):
    print("Recharge Successful!")
    break
  else:
    print("Please try again.\n")
```

### 2) Inventory Reorder Alert System

```python
# Inventory Reorder Alert System Prints Reorder Alert if stock < 15

# Function to check stock levels
def check_inventory(products):

    # Loop through dictionary items
    for product, stock in products.items():
        if stock < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")

# Sample product data
products = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 30
}

check_inventory(products)
```

### 3) Student Result Processing System

```python
# Student Result Processing System Calculates average marks and returns Pass/Fail.

# Function to process student result
def process_result(marks):
    total = 0

    # Loop to calculate total
    for mark in marks:
        total += mark
    # Calculate average
    average = total / len(marks)

    print("Average Marks:", average)
    # Return result based on average
    if average >= 50:
        return "Pass"
    else:
        return "Fail"


marks = [60, 55, 70, 45]
result = process_result(marks)
print("Result:", result)
```

### 4) Cab Fare Estimator with Retry Option

```python
# Cab Fare Estimator Base fare = ₹50 ₹12 per km Peak hour → 25% extra Allows retry.

# Function to calculate fare
def calculate_fare(distance, peak):
    fare = 50 + (12 * distance) # Calculation of base + per km calculation

    # Add 25% extra if peak hour
    if peak.lower() == "yes":
        fare += fare * 0.25

    return fare

# loop for continuing another calculation if user want
while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ")

    total_fare = calculate_fare(distance, peak)

    print("Total Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```

### 5) Employee Attendance Eligibility Checker

```python
# Employee Attendance Eligibility Checker Returns Eligible if attendance ≥ 75%

# Function to check eligibility
def check_eligibility(attendance):
    present_days = 0
    # Count present days
    for day in attendance:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance)) * 100

    print("Attendance Percentage:", percentage)

    # Return eligibility status
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance = ["P", "P", "A", "P", "P"]
status = check_eligibility(attendance)
print("Status:", status)
```

### 6) Password Strength Checker

```python
# Password Strength Checker Minimum length = 8 Must contain at least one digit Must contain at least one special character (@#$)

# Function to check password strength
def check_password(password):

    # Check minimum length
    if len(password) < 8:
        return "Weak Password (Too Short)"

    has_digit = False
    has_special = False
    special_chars = "@#$"
    # Loop through each character
    for char in password:

         # Check if character is digit
        if char.isdigit():
            has_digit = True

        # Check for special character
        if char in special_chars:
            has_special = True

    # validation of password
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


password = input("Enter password: ")
result = check_password(password)
print(result)
```
