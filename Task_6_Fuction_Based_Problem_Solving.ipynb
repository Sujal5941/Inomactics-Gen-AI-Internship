### 1) Smart Parking Lot Management System.

```python
# This program Design to manage a vehical entry and exit logs
# Description:
# Accept Vechile entry and exit logs.
# Calculate total parked vehicles.
# identify peak parking usage.
# Alert if parking exceeds capacity.

def manage_parking(vehical_logs):
  current_vehical_park = 0
  peak_usage = 0

  # for loop through vehical logs
  for log in vehical_logs:
    if log == "IN":    # if vehical IN current parking +1
      current_vehical_park  += 1
    elif log == "OUT":      # if vehical OUT current parking -1
      if current_vehical_park > 0:      # Prevent the negative parking count
        current_vehical_park -= 1

    if current_vehical_park > peak_usage:   # calculate the peak usage
      peak_usage = current_vehical_park

  # Displaying the current parked, peak usage and is Parking limit exceed or parking Available
  print("Currently Parked Vehicles : ", current_vehical_park)
  print("Peak Usage : ",peak_usage)
  if current_vehical_park > parking_capacity:
    print("Parking exceed Limit !!!!")
  else:
    print("Parking Status: Available")



parking_capacity = 50
vehical_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

manage_parking(vehical_logs)
```

### 2) Online Food Delivery Time Estimator

```python
# This program demonstrate the function that estimates delivery time based on : Distance(km), Weather condition, Traffic level.

def delivery_time_calculator(distance, weather, traffic):
  delivery_time = 0
  delivery_time = distance * 3  # require 3 min for trvelling 1 km

  if traffic == "high":   # if traffic is high delivery time increase by 50%
    delivery_time += delivery_time * 0.5
  elif traffic == "medium":   # if traffic is medium delivery time increase by 20%
    delivery_time += delivery_time * 0.2
  elif traffic == "low":      # if traffic is low delivery time has no change
    delivery_time += delivery_time * 0

  if weather == "rainy":    # if weather is rainy delivery time increase by 30%
    delivery_time += delivery_time * 0.3
  elif weather == "windy":    # if weather is rainy delivery time increase by 20%
    delivery_time += delivery_time * 0.2
  elif weather == "sunny":    # if weather is sunny delivery time has no change
    delivery_time += delivery_time * 0

  return delivery_time

distance = int(input("Distance(km) : "))
weather = input("Weather (sunny/windy/rainy) : ").lower()
traffic = input("Traffic (high/medium/low) : ").lower()

delivery_time = delivery_time_calculator(distance,weather,traffic)
print(f"Estimated Delivery Time : {int(delivery_time)} minutes")
```

### 3) Movie Theatre Seat Occupancy

```python
# This program demonstrate the function that analyzes seat booking data
# Calculates occupancy percentage
# Determines if show is Housefull
# Suggests opening additional shows

def analyze_occupancy(total_seats, booked_seats):

    booked_count = 0        # Count number of booked seats

    for seat in booked_seats:   # for loop through booked seats
        if seat == 1:
            booked_count += 1

    occupancy = (booked_count / total_seats) * 100     # Calculate occupancy percentage

    print(f"Occupancy: {occupancy:.0f}%")

    # Determine show status
    if occupancy == 100:            # if occupancy is 100 status -> housefull
        print("Show Status: Housefull")
        print("Suggestion: Consider opening additional shows")
    elif occupancy >= 75:           # if occupancy is 75 status -> Almost Full
        print("Show Status: Almost Full")
    else:                           # occupancy < 75 status -> Seats Available
        print("Show Status: Seats Available")

total_seats = 200

booked_seats = [1] * 150

analyze_occupancy(total_seats, booked_seats)
```

```python
# This program demostrate the function to classify server load based on CPU usage readings.\
# Avergae CPU < 50 % -> Normal
# 50% - 80% -> Warning
# 80% -> Critical

def classify_server_load(cpu_readings):
    total = 0

    for reading in cpu_readings:    # Calculate total CPU usage
        total += reading

    avg = total / len(cpu_readings)     # Calculate average CPU usage

    print(f"Average CPU Load: {average:.0f}%")

    # Classify server load
    if avg < 50:          # if avg < 50 -> Status : Normal
        print("Server Status: Normal")
    elif avg <= 80:       # if avg <= 80 -> status : Warning
        print("Server Status: Warning")
    else:                # if avg > 80 status : Critical
        print("Server Status: Critical")


# Sample Input
cpu_data = [45, 60, 70, 85, 90]
classify_server_load(cpu_data)
```

### 5) Smart Classroom Resource Usage Monitor

```python
# This Program demonstrate the function that tracks usage of classroom resources(projector, AC, lights) and identifies overuse patterns.

def resource_usage_monitor(resource_usage):
    overused_resources = []

    threshold = 7      # Define threshold for overuse (example: more than 8 hours)

    # Check each resource usage
    for resource, hours in resource_usage.items():
        if hours > threshold:
            overused_resources.append(resource)

    if overused_resources:    # if overused_resource contain some items print overused
        print(f"Overused Resources: ", ", ".join(overused_resources))
        print("Energy Alert: Yes")
    else:                    # if overused_resource is empty print no overused
        print("Overused Resources: None")
        print("Energy Alert: No")

usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_usage_monitor(usage_data)
```

### 6) Online Event Registration Capacity Controller

```python
# This program deomstrate the function that manages event registrations.
# 1) Confirming within capacity
# 2) Preventing overbooking
# 3) Activating waitlist mode

def manage_registrations(capacity, total_registrations):

    confirmed = 0
    waitlisted = 0

    # Loop through each registration
    for i in range(total_registrations):
        if confirmed < capacity:    # Check if there is capacity
            confirmed += 1
        else:                       # If no capacity add to waitlist
            waitlisted += 1

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    if waitlisted > 0:        # if  waitlist is greater than 0 print Registration closed
        print("Registration Status: Closed")
    else:                     # if waitlist is 0 or less print registration closed
        print("Registration Status: Open")


capacity = int(input("Capacity : "))
registrations = int(input("Registrations : "))

manage_registrations(capacity, registrations)
```
