### 1) User Login Check

```python
# This code check user credential with the entered username name and password

def userlogin_check(input_username, input_password):
  
  # Predefined credentials
  username = "admin"
  password = "1234"

  # Checking Credentials 
  if input_username == username and input_password == password:
    return "Login Successful"
  else:
    return "Invalid Credentials"
  
# Taking input from user
username = input("Enter Username : ")
password = input("Enter Password : ")

result = userlogin_check(username, password)
print(result)
```

### 2) Pass / Fail Analyzer

```python
# This program analyzes student marks and calculates the total number of passed and failed students

def result_analyzer(marks):
  
  #Initial pass and fail count to 0
  pass_count = 0
  fail_count = 0

  # Checking each student's marks
  for m in marks:
    if m >= 50:
      pass_count += 1
    else:
      fail_count += 1
  return pass_count, fail_count

marks = [45, 78, 90, 33, 60]

passed,failed = result_analyzer(marks)

print("Total Student : ",len(marks))
print("Passed Student : ",passed)
print("Failed Student : ",failed)

```

### 3) Simple Data Cleaner

```python
# This program clean user names by removing extra spaces and converting them to lowercase

def data_cleaner(names):
  cleaned_list = []

  # Cleaning each name
  for n in names:
    cleaned_list.append(n.lower())

  return cleaned_list


names = ["Alice", "bob", "CHARLIE"]

cleaned_data = data_cleaner(names)

print("Original Names : ",names)
print("Cleaned Names : ",cleaned_data)
```

### 4) Message Length Analyzer

```python
# This program calculates the length of each message and flags messages longer than 10

def message_analyzer(messages):

  for m in messages:
    length = len(m)
    print(f"Message : '{m}' | Length : {length}")

    # Flag messages longer than 10 characters
    if length > 10:
      print("!!!!This Message is longer than 10 characters!!!!")
    print()

messages = ["Hi", "Welcome to the platform", "OK"]

message_analyzer(messages)
```

### 5) Error Message Detector

```python
# This program counts the total number of "ERROR" entries in a system log list

def Error_message_detector(logs):
  error_count = 0

  # Counting ERROR logs
  for l in logs:
    if l == "ERROR":
      error_count += 1

  return error_count


logs = ["INFO", "ERROR", "WARNING", "ERROR"]

total_error = Error_message_detector(logs)

print("Total ERROR entries : ",total_error)
```
