## **STEP 3: Data Understanding**

```python
print(df.shape)
print(df.columns)
print(df['sentiment'].value_counts())
```

## **Step 4: NLP Preprocessing**

```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

nltk.download('punkt')
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

def preprocess_text(text):
    # 1. Lowercase
    text = text.lower()

    # 2. Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # 3. Tokenization
    tokens = word_tokenize(text)

    # 4. Remove stopwords
    tokens = [word for word in tokens if word not in stop_words]

    # 5. Stemming
    tokens = [stemmer.stem(word) for word in tokens]

    return " ".join(tokens)
```

```python
df['clean_text'] = df['review'].apply(preprocess_text)
```

## **STEP 5: Feature Engineering**

#### **1. Bag of Words**

```python
from sklearn.feature_extraction.text import CountVectorizer

bow = CountVectorizer()
X_bow = bow.fit_transform(df['clean_text'])
```

#### **2. TF-IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
X_tfidf = tfidf.fit_transform(df['clean_text'])
```

## **STEP 6: Train-Test Split**

```python
from sklearn.model_selection import train_test_split

y = df['sentiment']

X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf, y, test_size=0.2, random_state=42
)
```

## **STEP 7: Model Building**

#### **1. Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)
```

**2. Naive Bayes**

```python
from sklearn.naive_bayes import MultinomialNB

nb = MultinomialNB()
nb.fit(X_train, y_train)

y_pred_nb = nb.predict(X_test)
```

#### **3. Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)
```

## **STEP 8: Model Evaluation**

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate_model(y_test, y_pred, model_name):
    print(f"--- {model_name} ---")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Precision:", precision_score(y_test, y_pred, average='weighted'))
    print("Recall:", recall_score(y_test, y_pred, average='weighted'))
    print("F1 Score:", f1_score(y_test, y_pred, average='weighted'))
    print()
```

```python
# Model Evaluation
evaluate_model(y_test, y_pred_lr, "Logistic Regression")
evaluate_model(y_test, y_pred_nb, "Naive Bayes")
evaluate_model(y_test, y_pred_dt, "Decision Tree")
```

## **Step 9: Comparison**

## Model Comparison & Insights

### Observations

- **Logistic Regression** achieved the highest accuracy (**89.23%**) among all models.  
  It performs well because it can handle high-dimensional sparse data generated by TF-IDF and works efficiently for linear classification problems.

- **Naive Bayes** also performed well with an accuracy of **86.41%**.  
  It is fast and efficient, but it assumes that all features are independent, which is not always true in real-world text data.

- **Decision Tree** showed the lowest performance (**71.59% accuracy**).  
  This is because Decision Trees are not well-suited for sparse and high-dimensional data and tend to overfit.

---

### Key Insights

- Logistic Regression is the **best-performing model** for this dataset.  
- Naive Bayes is a good **baseline model** due to its speed and simplicity.  
- Decision Tree is **not ideal for NLP tasks** with vectorized text data.

---

### Trade-offs

- Logistic Regression → High accuracy but requires proper feature engineering  
- Naive Bayes → Fast but based on strong assumptions  
- Decision Tree → Easy to understand but prone to overfitting
