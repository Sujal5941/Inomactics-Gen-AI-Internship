### **Task 2:Build Advanced Preprocessing Function**

```python
import re

def preprocess_text(text):
    try:
        if not text or text.strip() == "":
            return [], ""

        # Remove URLs
        text = re.sub(r'http\S+|www\S+', '', text)

        # Remove emails
        text = re.sub(r'\S+@\S+', '', text)

        # Remove numbers
        text = re.sub(r'\d+', '', text)

        # Convert to lowercase
        text = text.lower()

        # Handle repeated characters (soooo → soo → so)
        text = re.sub(r'(.)\1{2,}', r'\1', text)

        # Remove special characters & emojis
        text = re.sub(r'[^a-z\s]', '', text)

        # Remove extra spaces
        text = re.sub(r'\s+', ' ', text).strip()

        tokens = text.split()

        # Remove short words except 'no', 'not'
        filtered_tokens = [
            word for word in tokens
            if len(word) > 2 or word in ['no', 'not']
        ]

        clean_sentence = " ".join(filtered_tokens)

        return filtered_tokens, clean_sentence

    except Exception as e:
        return [], ""
```

```python
edge_cases = [
    "",                # empty string
    "😍😍🔥🔥",        # only emojis
    "1234567890",      # only numbers
    None               # invalid input
]

for i, text in enumerate(edge_cases, 1):
    result = preprocess_text(text)

    print("="*50)
    print(f"Test Case {i}")
    print("Input :", text)
    print("Output:", result)
```

```python
text = "WOW!!! This is soooo goooood!!! I have 2 dogs. Visit http://example.com now or mail me at test@gmail.com"

result = preprocess_text(text)

print(result)
```

### **Task 3: Stress Testing**

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens, clean = preprocess_text(sentence)

    print("\nOriginal:", sentence)
    print("Tokens:", tokens)
    print("Clean Sentence:", clean)

    results.append(tokens)
```

### **Task 4: Token Analytics**

```python
import numpy as np

for i, tokens in enumerate(results):
    if len(tokens) == 0:
        print(f"\nSentence {i+1}: No valid tokens")
        continue

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = np.mean([len(word) for word in tokens])

    print(f"\nSentence {i+1}:")
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", round(avg_len, 2))
```

### **Task 5: Frequency Analysis**

```python
from collections import Counter

all_tokens = [word for sentence in results for word in sentence]

counter = Counter(all_tokens)

print("\nTop 10 Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

### **Task 6: Build Full Pipeline**

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
result = full_pipeline(test_sentences)

print("Clean Sentences:")
print(result["clean_sentences"])

print("\nAll Tokens:")
print(result["tokens"])
```
