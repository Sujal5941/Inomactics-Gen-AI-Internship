## **Prompt Templates using LangChain**

### **Install & Imports**

```python
# Install (already installed in your case)
!pip install langchain

# Correct Imports
from langchain_core.prompts import PromptTemplate
from langchain_core.prompts import ChatPromptTemplate

print("LangChain imported successfully")
```

### **Task 1: Replace Hardcoded Prompt**

```python
template_1 = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

# Test
print(template_1.format(topic="Artificial Intelligence"))
```

### **Task 2: Multi-Input Prompt System**

```python
multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# Test Cases
print(multi_input_template.format(topic="AI", audience="beginners", tone="friendly"))
print(multi_input_template.format(topic="Python", audience="kids", tone="fun"))
print(multi_input_template.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

### **Task 3: Prompt Variations Engine**

```python
# Teaching Template
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

# Interview Template
interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

# Storytelling Template
story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

# Run for Machine Learning
topic = "Machine Learning"

print(teaching_template.format(topic=topic))
print(interview_template.format(topic=topic))
print(story_template.format(topic=topic))
```

### **Task 4: ChatPromptTemplate System**

```python
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful {role}"),
    ("user", "Explain {topic}")
])

# Example
messages = chat_template.format_messages(role="teacher", topic="Neural Networks")

for msg in messages:
    print(msg)
```

### **Task 5: Input Validation Layer**

```python
def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    # Default values
    if audience not in valid_audience:
        print("Invalid audience → setting to 'beginner'")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone → setting to 'formal'")
        tone = "formal"

    return audience, tone

# Test
audience, tone = validate_inputs("kids", "funny")
print(audience, tone)
```

### **Task 6: Prompt Generator App**

```python
def generate_prompt(topic, audience, tone, style):

    audience, tone = validate_inputs(audience, tone)

    # Style templates
    if style == "teaching":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Teach {topic} to {audience} in a {tone} tone"
        )

    elif style == "interview":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Generate interview questions on {topic} for {audience} in a {tone} tone"
        )

    elif style == "storytelling":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} as a story for {audience} in a {tone} tone"
        )

    else:
        return "Invalid style"

    return template.format(topic=topic, audience=audience, tone=tone)

# Test
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

### **Task 7: Template Reusability Test**

```python
reusable_template = PromptTemplate(
    input_variables=["topic", "audience"],
    template="Explain {topic} to {audience} in simple terms"
)

# 5 different inputs
print(reusable_template.format(topic="AI", audience="kids"))
print(reusable_template.format(topic="Blockchain", audience="beginners"))
print(reusable_template.format(topic="Cloud Computing", audience="students"))
print(reusable_template.format(topic="Data Science", audience="analysts"))
print(reusable_template.format(topic="Cybersecurity", audience="professionals"))
```

### **Final Pipeline Flow**
User Input → Validation → Prompt Template → Dynamic Prompt → Output

```python
# Bonus: Full pipeline test
topic = "Artificial Intelligence"
audience = "beginner"
tone = "fun"
style = "teaching"

output = generate_prompt(topic, audience, tone, style)
print("Final Generated Prompt:\n", output)
```
