### 1) Employee Performance Bonus Eligibility

```python
# This program identifies employeeseligible for top performance bonus.
# It finds the highest performance score from a dictionary of employees and handles tie cases where multiple employees share the same top score.

employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}

highest_score = 0

# Identify the highest performance score
for emp in employees:
  score = employees[emp]
  if score > highest_score:
    highest_score = score

# handle multiple top performers
top_performers = []

#
for name,score in employees.items():
  if score == highest_score:
    top_performers.append(name)

print(f"Top performer eligible for Bonus : {top_performers} , Score:{highest_score}")

```

### 2) Search Query Keyword Analysis


```python
# This program analyzes a customer search query.
# It converts the input to lowercase, removes punctuation, counts keyword frequency, and displays only keywords that appear more than once,

query = "Buy mobile phone buy, phone online"

# Convert to lowercase
query =  query.lower()

# Remove punctuation
import string
for char in string.punctuation:
  query = query.replace(char, "")

# Split into words
words = query.split()

# Counting frequency
frequency={}

for word in words:
  if word in frequency:
    frequency[word] += 1
  else:
    frequency[word] = 1

# Display words search more than one
result = {}

for word in frequency:
  if frequency[word] > 1:
    result[word] = frequency[word]

print(result)
```

### 3) Sensor Data Validation

```python
# This program validates sensor readings collected hourly.
# It identifies even-numbered readings (valid readings) and stores them as (hour_index, reading_value) pairs.Odd readings are ignored

sensor_readings=[3, 4, 7, 8, 10, 12, 5]

valid_readings =[]

# storing the value in the form of (index,value) in the valid_readings
for index,value in enumerate(sensor_readings):
  if value % 2 == 0:
    # adding even number reading into valid_readings
    valid_readings.append((index,value))

print("Valid Sensor Readings (Hour, Value) : ",valid_readings)
```

### 4) Email Domain Usage Analysis

```python
# This program analyzes employee email IDs to determine how many users belong to each email
# and calculates the percentage usage of each domain

emails =[
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# storing the number of domain count in the form of domain name and no of counts
domain_count = {}

# accessing emails
for email in emails:
  #spliting the string
  domain = email.split("@")[1]
  # counting the domain count
  domain_count[domain] = domain_count.get(domain,0) + 1

total_users = len(emails)

# calculating the percentage of each domain
for domain,count in domain_count.items():
  percentage = (count/ total_users) * 100
  print(f"{domain} : {percentage:.0f}% ")
```

### 5) Sales Spike Detection

```python
# This program calculates the average daily sales and detects days where sales exceed 30% above avg value

sales = [1200, 1500, 900, 2200, 1400, 3000]

# taking avg of slaes
average_sales = sum(sales) / len(sales)
# calculating threshold for the sales exceed 30% above avg value
threshold = (average_sales) * 1.30

# displaying the day at which the value exceed 30% above avg value (threshold)
for day, value in enumerate(sales, start=1):
  if value > threshold:
    print(f"Day {day} : {value}")
```

### 6) Duplicate User ID Detection

```python
# This program identifies duplicates user IDs and displays how many times each duplicate appears

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# storing number of duplicate IDs
frequency = {}

# counting the number of user
for user in user_ids:
  frequency[user] = frequency.get(user, 0) + 1

# displaying the number of duplicate user
for user in frequency:
  if frequency[user] > 1:
    print(f"{user} -> {frequency[user]} times")
```
