### 1) Unique Words in a Sentence

```python
# This program identifies unique words from a sentence and displays the count along with the unique words.

sentence = input("Enter sentcnce : ")

# Split the sentence into words
words = sentence.split()

# Convert list to set to remove duplicates
unique_words = set(words)

print("Unique words count : ",len(unique_words))
print("Unique words : ",unique_words)
```

### 2) Highest Salary from Employee Data

```python
# This program finds the employee with the highest salary from a dictionary of employees.

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_employee = None
highest_salary = 0

for name,salary in employees.items():
  # compare the highest salary of employee
  if salary > highest_salary:
    highest_salary = salary
    highest_employee = name

print("Highest salaray : ",highest_employee, "-", highest_salary)
```

### 3) Find Maximum and Minimum Values

```python
# This program finds the maximum and minimum values from a list of numbers.

numbers = [45, 22, 89, 10, 66]

max_num = numbers[0]
min_num = numbers[0]

# for loop for accessing each element in list
for num in numbers:
  # comparing num is max
  if num > max_num:
    max_num = num
  # comparing num is min
  if num < min_num:
    min_num = num

print("List : ",numbers)
print("Max : ",max_num)
print("Min : ",min_num)
```

### 4) Count Products Above a Price Threshold

```python
# This program counts how many product prices are greater than 1000.

prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
  # comparing the price greater than 1000
  if price > 1000:
    count += 1

print("Products above 1000 : ",count)
```

### 5) Calculate Attendance Percentage

```python
# This program calculates attendance percentage based on "P" (Present) and "A"(Absent).

attendance = ["P","P","A","P","P"]

present_count = 0

# for loop for present count
for a in attendance:
  if a  == "P":
    present_count += 1
total_days = len(attendance)

percentage = (present_count / total_days) * 100
print("Attendance Percentage : ",percentage)
```

### 7) Count Character Frequency

```python
# This program counts how many times each character appears in a string.

text = "pythonp"

frequency = {}

for char in text:
  # check how many times character appear in string
  if char in frequency:
    frequency[char] += 1
  else:
    frequency[char] = 1

print(frequency)
```
