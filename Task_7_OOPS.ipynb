### **1) Smart Light Controller:**



```python
# This program create a class to control a smart light that can br turned ON or OFF and display its status

class SmartLight:     # define a class smartlight
  def __init__(self,lightname,action):
    self.lightname = lightname
    self.action = action

  def controller(self,action):    # define a controller method to controller and display the status
    if action == "ON":        # if action is "ON" then these block execute
      print(f"{self.lightname} is ON")
    elif action == "OFF":     # elif action is "OFF" then these block execute
      print(f"{self.lightname} is OFF")
    else:     # else action is something random then these block execute
      print("Invalid action")

lightname = input("Light Name : ")
action = input("Action : ").upper()

samrtlight = SmartLight(lightname,action)
samrtlight.controller(action)
```

### **2) Employee ID Card System**

```python
# Create a class that stores employee name,ID, and department and displays ID card details.

class Employee:     # define a class Employee
  def __init__(self,name,Id,department):
    self.name = name
    self.Id = Id
    self.department = department
  def display_IDCard(self):     # define a method display_IDCard
    print("Employee ID Card")
    print(f"Name : {self.name}")
    print(f"ID : {self.Id}")
    print(f"Department : {self.department}")

name = input("Employee Name :")
Id = input("Employee ID : ")
department = input("Employee Department : ")

emp = Employee(name,Id,department)      # creating an object of a class Employee
emp.display_IDCard()        # calling method display_IDCard from class Employee
```

### **3) Mobile Contact Record**

```python
# Create a class to store contact name and phone number and display contact information.

class ContactRecord:      # define a class ContactRecord
  def __init__(self,name,phoneNo):
    self.name = name
    self.phoneNo = phoneNo

  def display_contact(self):     # define a method display_contact
    print("Contact Saved")
    print(f"Name : {self.name}")
    print(f"Phone Number : {self.phoneNo}")

name = input("Contact Name : ")
phoneNo = input("Phone NUmber : ")

contact = ContactRecord(name, phoneNo)      # creating an object of a class ContactRecord
contact.display_contact()       # calling method display_contact from class ContactRecord
```

### **4) Product Price Tag Generator**

```python
# Create a class that stores product name and price and prints a formatted price tag.

class Product:        # define a class Product
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_price_tag(self):      # define a method display_price_tag
        print("Product:", self.name)
        print("Price: ₹",self.price)

name = input("Product Name: ")
price = input("Price: ")

product = Product(name, price)      # creating an object of a class Product
product.display_price_tag()     # calling method display_price_tag from class Product
```

### **Movie Rating Display System**

```python
# This program create a class that stores movie name and rating and displays movie details.

class Movie:      # define a class Movie
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie(self):        # define a method display_movie
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")

name = input("Movie Name: ")
rating = input("Rating: ")

movie = Movie(name, rating)     # creating an object of a class Movie
movie.display_movie()       # calling method display_movie from class Movie
```

### **6) Delivery Address Manager**

```python
# This program create a class that stores customer name and delivery address and prints delivery details.

class Delivery:       # define a class Delivery
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display_delivery(self):       # define a method display_delivery
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)

name = input("Customer Name: ")
address = input("Address: ")

delivery = Delivery(name, address)      # creating an object of a class Delivery
delivery.display_delivery()       # calling method display_delivery from class Delivery
```
