### 1) Social Media – Post Engagement Analyzer

```python
# This program demonstrate the Social Media Post Engagement Analyzer.
# Which Calculates the total_likes and analyze the post is viral post if likes >= 1000 else Normal Engagement

def analyze_engagement(Likes):
  total_likes= 0

  # Calculates the total likes
  for likes in Likes:
    total_likes += likes

  print("Total Likes : ",total_likes)

  # if-else for like is >= 1000 it will print 'Viral Post' if not then 'Normal Engagement'
  if total_likes >= 1000:
    print("Post Status : Viral Post")
  else:
    print("Post Status : Normal Engagement")

likes = [200,400,350,450,150]
analyze_engagement(likes)
```

### 3) Agricultural Rainfall Adequacy Checker

```python
# This program demonstrate the function that analyzes seasonal rainfall data
# calculate the avg rainfall, compairing the avg rainfall with the required rainfall if greater then it is adequate else inadequate

def rainfall_checker(rainfall, required_rainfall):
  total_rainfall = 0

  # for loop for calculationg total rain
  for rain in rainfall:
    total_rainfall += rain

  # calculate the avg rainfall
  avg_rainfall = total_rainfall/len(rainfall)
  print("Average Rainfall: ",avg_rainfall)

  # if-else for checking rainfall is adequate or inadequate
  if avg_rainfall >= required_rainfall:
    print("Rainfall Status : Adequate Rainfall")
  else:
    print("Rainfall Status : Inadequate Rainfall")

rainfall = [76,62,77,65,72]
required_rainfall = 65
rainfall_checker(rainfall, required_rainfall)
```

### 6) Agriculture - Premium Crop Price Filter

```python
# This program demonstrate the function that filters crop above a premium threshold.
# If crop price of crop > ₹2000 'Premium Crop' else 'Regular Crop'

def crop_filter(prices):
  premium = []
  # for loop goes through list of prices
  for price in prices:
    # if-else for checking crop is premium or regular
    if price > 2000:
      premium.append(price)
  return premium

prices = [1500,1200,2200,1800,2500,2500,3200,1000]
print("Premium Crops : ",crop_filter(prices))
```

### 7) System Monitoring - Application Health Checker

```python
# This program demonstrate the function to evaluate application health based on error count
# if errors = 0 -> healthy, errors <= 5 -> Minor Issues, Errors > 5 -> Critical issue

def health_checker(errors):
  # if-elif-else to check the system status.
  if errors == 0:
    print("System Status : Healthy")
  elif errors <= 5:
    print("System Status : Minor Issues")
  else:
    print("System Status : Critical Issues")

errors = int(input("Error Count : "))
health_checker(errors)
```

### 8) Banking – Daily Transaction Limit Checker

```python
# This program demonstrate the function to validate daily transaction amount.
# Daily limit of account is 50000, if amount within limit -> Approved else -> Rejected

def Transaction_checker(amount):
  daily_limit = 50000

  # If amount < 0 -> 'Invalid Amount', else amount <= daily limit -> 'Approved', else 'Rejected'.
  if amount <0:
    print("Transaction Status : Invalid Amount")
  elif amount <= daily_limit:
    print("Transaction Status : Approved")
  else:
    print("Transaction Status : Rejected")

amount = int(input("Transaction Amount : "))
Transaction_checker(amount)
```

### 9) E-Learning - Student Attendance Eligibility System


```python
# This program demonstrate the function to calculate the percentage and check exaam eligibility.
# If Attendance >= 75% -> 'Eligible' else Attendance < 75% -> 'Not Eligible'

def eligibility_system(attendance):
  attendance_count = 0    # initialize attendance_count
  # for loop through attendance
  for attend in attendance:
    if attend == "P":     # if attend then attendance_count ++
      attendance_count += 1

  # Calculate attendance percentage
  attendance_precentage = (attendance_count/len(attendance))*100
  print("Attendance Percenatge : ",attendance_precentage)

  # If attendance_percentage >= 75 print 'Eligible' else 'Not Eligible'
  if attendance_precentage >= 75:
    print("Exam Eligibility : Eligible")
  else:
    print("Exam Eligibility : Not Eligible")

attendance = ["P","P","A","P","P","A","P","P","P","P"]
eligibility_system(attendance)
```

### 10) Smart Electricity Bill Analyzer


```python
# Problem Statement : Smart electric billing system for residential users.
# This program demonstrate the function that calculates the electric bill and classifies usage level based on units consumed.
# Billing rules => 1st 100 units -> ₹3 per unit, next 100 units(101-200)-> ₹5 per unit, above 200 units -> ₹7 per unit.
# Usage Classification => if total bill < ₹500 'low usage', if between(₹500-₹1500) 'Moderate Usage' and higher than ₹1500 'High Usage'.

def bill_calculator(units):   # function for bill calculation
  if units <=0:
    return 0
  elif units <= 100:
    return units * 3
  elif units <= 200:
    return (100 * 3)+((units-100) * 5)
  else:
    return (100 * 3)+(100 * 5)+((units-200) * 7)

def usage_classification(bill):   # function for usage classification
  if bill < 500:
    return "Low Usage"
  elif bill <=1500:
    return "Moderate Usage"
  else:
    return "High Usage"

units = int(input("Enter Consumed Units : "))
total_bill = bill_calculator(units)
print("Total bill = ",total_bill)
print("Usage Classification : ",usage_classification(total_bill))
```
